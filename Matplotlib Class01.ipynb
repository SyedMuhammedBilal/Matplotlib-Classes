# FIRST PLOTS WITH MATPLOTLIB

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot([1,2,3,4,5],[2,4,6,8,10])
plt.grid(True)
plt.show()
```

```python
x = range(6)
plt.plot(x,[y**2 for y in x])
plt.show()
```

```python
x = np.arange(0.0, 6.0, 0.01)
plt.plot(x, [x**2 for x in x])
plt.show()
```

```python
x = range(1,5)
plt.plot(x, [y*1.5 for y in x])
plt.plot(x, [y*15 for y in x])
plt.plot(x, [y*7.5 for y in x])

plt.show()
```

# MULTILINE PLOTS

```python
x = range(1,5)
plt.plot(x, [xi*1.7 for xi in x], x, [xi*3.0 for xi in x], x,[xi/9.0 for xi in x])
plt.grid(True)
plt.show()
```

```python
x = np.arange(0, 50, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.grid(True)
plt.ylabel('Y-axis')
plt.xlabel('X-axis')
plt.title('FIRST GRAPH')
plt.show()
```

# ADDING LEGEND

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1, 5)
plt.plot(x, x*1.5, label='Normal')
plt.plot(x, x*3.0, label='Fast')
plt.plot(x, x/3.0, label='Slow')
plt.legend()
plt.show()
```

```python
x = np.arange(0, 120, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.grid(True, color='black')
plt.ylabel('Y-axis', color='green')
plt.xlabel('X-axis', color='green')
plt.title('FIRST GRAPH', color='blue')
plt.plot(x,y, c='red')
plt.show()
```

# COMPLETE EXAMPLE

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1, 5)
plt.plot(x, x*1.5, label='Normal')
plt.plot(x, x*3.0, label='Fast')
plt.plot(x, x/3.0, label='Slow')
plt.grid(True)
plt.title('Sample Growth of a Measure')
plt.xlabel('Samples')
plt.ylabel('Values Measured')
plt.legend(loc='upper left')
plt.savefig('plot123.png')
plt.show()
```

# COLOUR CONTROL

```python
y = np.arange(1,3)
plt.plot(y, 'y')
plt.plot(y+1 , 'm')
plt.plot(y+2 , 'c')
plt.show()
```

```python
y=np.arange(1,3)
plt.plot(y,'y',y+1,'m',y+2,'#0200a8') #hex color picker from google to choose any color
                                #data,color,data,color
plt.show()
```

# LINE STYLES CONTROL

```python
y = np.arange(1,3)
plt.plot(y, '--', y+1, '-.', y+2, ':')
plt.show()
```

# CONTROL MARKER STYLES

```python
y = np.arange(1, 3, 0.2)
plt.plot(y, 'x', y+0.5, 'o', y+1, 'D', y+1.5, '^', y+2, 's')
plt.grid(True)
plt.show()
```

```python
y = np.arange(1, 3, 0.3)
plt.plot(y, color='blue', linestyle='dashdot', linewidth=4, marker='o', markerfacecolor='red', markeredgecolor='black', markeredgewidth=3, markersize=12)
plt.show()
```

# HANDLING X & Y TICKS

```python
import matplotlib.pyplot as plt
x = [5, 3, 7, 2, 4, 1]
plt.plot(x);
plt.xticks(range(len(x)), ['a', 'b', 'c', 'd', 'e', 'f']);
plt.yticks(range(1, 8, 2));
plt.plot(x, c='red')
plt.show()
```

# BAR GRAPH WITH TICKS-CHARTS
